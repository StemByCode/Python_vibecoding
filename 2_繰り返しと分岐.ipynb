# 2. 繰り返しと分岐

Pythonのデータ処理の手順は、プログラムを上から読んでいき（順次実行）、条件による分岐があれば処理が分岐し、繰り返しの命令があれば処理を繰り返す、というのが基本です。

例えば、前回習ったリストの形で入力データが与えられ、処理（順次実行、分岐、繰り返し）が行われ、整数が出力される、という感じ。

## 2-1. 繰り返し処理（forとwhile）

例題2-1-1. 35から45までの数字をすべて足し合わせ、総和を出力してください。

```python
tot = 0
for i in range(35,46):  # for ～ in ～は定型文です。
    tot += i
print(f"総和は{tot}です。")
```

例題2-1-1のように『何をいつからいつまで繰り返すか』というのが明らかな場合はfor文を使います。

ここで注意しないといけないのが**インデント**です。  
インデントというのは字下げのことで、ここでは半角４つ分さげることを指します。
文字の開始をインデントすることで、ある処理（例題2-1-1の場合はfor文）の有効範囲がどこからどこまでなのかが目で見てわかります。  
なお、for文の中にfor文があれば、内側のfor文でさらにインデントが発生することになります。  
ちなみに、google colaboratoryの場合、ツール→設定→エディタ→インデント幅でインデントの幅を調整できます。

例題2-1-2. 数列$a_{n+1}=2a_n+3$, $a_0=0$を考えます。$a_n\ge 10000$を満足する最小の$a_n$を求めてください。

```python
a = 0  # 初期値
while a<10000:  # aが1000より小さい間は処理を続ける
  a = 2*a+3  # 次の項
print(a)
```

例題2-1-2のように『いつ終わるか明確にわからないけれども、終わるための条件がわかっている』という場合にはwhile文を使います。

## 2-2.分岐処理（if文）

if～elif～elseに慣れましょう。  
例題は定番のFizz Buzzで。

例題2-2-1. 10から50までの数字を順番に出力してください。ただし、i番目の数字をaとするとき、aが3で割り切れるならばaではなく"Fizz"と、aが5で割り切れるならばaではなくて"Buzz"と、aが3でも5でも割り切れるならばaではなくて"FizzBuzz"と出力すること。

```python
# 問題文もわざとわかりにくくしている感じですが。。。
# 処理が上から順番に実行されていくのを理解してください。
for i in range(10,51):
  if i%3==0 and i%5==0:
    print("FizzBuzz")
  elif i%3==0:
    print("Fizz")
  elif i%5==0:
    print("Buzz")
  else:
    print(i)
```

今回は制御文（分岐と繰り返し）を習いました。前回のリストと合わせると実はほとんど何でもできるようになっています。
