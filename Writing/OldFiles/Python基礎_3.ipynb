## 4. tuple, dict, set

Pythonのデータ構造は、list（リスト）が最も基本と言えますが、基本的なデータ構造は他にもあります。それがタイトルにあるtuple（タプル）とdict（辞書）とset（集合）です。

### 4-1. tuple

tupleは一言でいえば、参照しかできないリストです。追加や削除はできません。  
「それならlistで十分じゃないの。」と思ったあなた。その通りです。  
が、tupleにはtupleの使い道があるのです。一言でいえば「dictのキーやsetの要素になり得る」ということです。listにはできません。  
tupleの例題は次のふたつ。

例題4-1. tupleをひとつ作り、その最後の要素を参照せよ。

```python
A = (3,4,5)
print(A[-1])
```

と、いうように丸括弧で囲ったものがtupleとなります。参照操作はlistと同じ。スライスも可能。  
１点注意すべきことは、見た目ただの丸括弧と区別がつかなくなるということです。  
例えば$(3*5)$と会った場合は、加減乗除の処理の途中でよくでてくる数字を括弧でくくったものなのか、それともtupleなのか。  
実は、上の$(3*5)$はPythonでは明確に前者である、と言うことができます。

例題4-2. 要素が$3*2$であるようなtupleを作ってください。

```python
A = (3*2,)
type(A)
```

というように、tupleの場合は数字の後にカンマを入れることでtupleだと主張します。

ほかの操作はlistと同じ。  
問題4-1. Aを要素が20から100までの整数が順に並んだtupleとする。Aの15番目の値を参照せよ。

### 4-2. dict

dictは例えばA = {"太郎":20, "次郎":10, "三郎":5}というように、{}の中に:（コロン）でつないだkeyとvalueをカンマで区切ったものです。  
上の例では、dictであるAのkeyは"太郎"、"次郎"、"三郎"で、各keyに対応するvalueは20、10、5ということになります。
このように何らかの対応を表現したいときはdictが便利です。  
keyになり得るのは上の例のように文字列や数やtupleですが、valueにはそのような制限がなく、listやsetもvalueになり得ます。
dictのkeyになり得るものとsetの要素になり得るものは同じということは頭の片隅に置いておくとよいですね。

例題4-3. dict を構成せよ。

```python
# keyの数が３つ程度ならば
A1 = {"太郎":20, "次郎":10, "三郎":5}  # そのまんま定義
print(A1)

# keyの数が多くなれば
A2 = {}  # 空のdict
for i in range(10):
    # keyとvalueを個別に設定する
    A2[str(i)+"-太郎"]=10+5*i  # str(i)は整数型の変数iを文字型に変換する関数
print(A2)
```

例題4-4. A1からkeyが次郎である要素を削除せよ。

```python
del(A1["次郎"])
A1
```

dictはいろいろ便利な関数（正確にはメソッド）がありますが、とりあえず次の例題のものは知っておいて損はない。

例題4-5. dict A2からkeyのみ取り出してlistとして構成せよ。

```python
B21 = list(A2.keys())
B21
```

例題4-6. dict A2からvalueのみ取り出してlistとして構成せよ。

```python
B22 = list(A2.values())
B22
```

例題4-7. dict A2から(key,value)のようにtupleとして取り出したものをlistとして構成せよ。

```python
B23 = list(A2.items())
B23
```

例えばdictとして構成されたものをリストにすることで、何らかの順番でソートが可能になります。

例題4-8. dict A1 = {"太郎":20, "次郎":10, "三郎":5} をlistにして、valueに関して昇順に並べ替えなさい。

```python
A1 = {"太郎":20, "次郎":10, "三郎":5} 
B1 = list(A1.items())  # list化
B1 = sorted(B1,key=lambda x:x[1])  # B1の要素(x[0],x[1])において、x[1]の要素について昇順にソート
B1
```

上の例題のsorted関数やlambda関数についてはまた別の機会に詳しくやります。

問題4-2. 上で定義したdict A2についてlist化し、valueに関して降順に並べ替えなさい。

### 4-3. set

例題4-9. 空集合を定義し、その集合に要素"cat"を追加せよ。

```python
A = set()  # 空集合
A.add("cat")
A
```

setの場合、要素の追加はaddを使います。  
また、波括弧はsetだけでなくdictでも使われているので、空dictは{}、空setはset()として区別します。

問題4-3. 20以上100未満の整数からなる集合を構成せよ。

ところで集合演算の基本は共通部分と和集合ですね。

例題4-10. set B = {"dog","cat","python"}を定義し、上の例題で作ったset Aとの共通部分と和集合を求めよ。

```python
B = {"dog","cat","python"}
C = A&B  # 共通部分
print(C)
D = A|B  # 和集合
print(D)
```

例題4-11. 上で作った集合Dから集合Aの要素を取り除いた集合Eを構成せよ。

```python
E = D-A  # 差集合
print(E)
```

setから要素を取り除くときはremoveを使います。これはlistの場合と同じ。
ちなみにdelは要素のインデックス（もしくはキー）を指定して削除するイメージ。

例題4-12. 上で作った集合Dから"python"を削除せよ。

```python
D.remove("python")
D
```

たぶん薄々きづいていることかもしれませんが、データ構造Aに対して働く関数にもf(A)とする関数や、A.f()とする関数（正確にはメソッドですが）の２種類があります。前者は元のデータA自体は変えずに、関数の戻り値としてデータAに操作を行ったものが与えられます。
後者はデータAに直接操作を行うもので、戻り値はありません。
前者の例としてはsorted()関数、後者の例としてはすぐ上の例題ででてきたremoveメソッドがあります。  
関数を使うときは、データ自体を直接変更するものかどうかということに注意を払いながら使うのが安全です。

### 4-4. 内包表記

人によって好き嫌いが分かれると思いますが、表題の内包表記はデータの定義の仕方に便利な方法です。

例題4-13. 0オリジンでi番目の要素がiであるような要素数１００のリストを構成せよ。

```python
# 普通の構成
A = []
for i in range(100):
    A.append(i)
# 内包表記
B = [i for i in range(100)]
```

要するにlistやtupleやdictやsetを定義する際に括弧の中でfor文を使って簡潔にすませることです。
コードがすっきりすることと実行速度が普通に構成するよりも若干速くなるのがメリット。
