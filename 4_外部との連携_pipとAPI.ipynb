## Information extraction

### Subtask:
Develop logic to process the search results and extract relevant information, specifically the locations (addresses or coordinates) of the accidents. This might involve string parsing or more advanced techniques if the data format is complex.


**Reasoning**:
Implement the function `extract_accident_locations` to parse the search results and extract potential accident locations based on keywords.



```python
import re

def extract_accident_locations(search_results):
    """
    Extracts potential accident locations from Tavily search results.

    Args:
        search_results (list): A list of dictionaries, where each dictionary
                                represents a search result from Tavily.

    Returns:
        list: A list of strings, where each string is a potential accident location.
    """
    locations = []
    location_keywords = ["場所", "付近", "交差点", "区", "市", "町", "村", "丁目", "番地", "地先", "道路"]

    for result in search_results:
        content = result.get('content', '')
        # Combine title and content for better parsing
        text_to_parse = f"{result.get('title', '')}. {content}"

        # Simple regex pattern to find potential locations near keywords
        # This is a basic approach and might need refinement based on actual search result formats
        pattern = r'(' + '|'.join(location_keywords) + r')\s*([\w\s\d-]+)'
        matches = re.findall(pattern, text_to_parse)

        for keyword, location_text in matches:
            # Basic cleaning and adding to the list
            cleaned_location = location_text.strip().replace('。', '').replace('、', '')
            if cleaned_location:
                locations.append(cleaned_location)
                # For now, we'll take the first few found locations per result to avoid noise
                if len(locations) > len(search_results) * 2: # Limit to a reasonable number
                    break
        if len(locations) > len(search_results) * 2:
            break

    # Remove duplicates and return
    return list(set(locations))

# Note: Actual search results format can vary greatly.
# This extraction logic is a starting point and may require
# significant refinement based on the real-world data from the API.
# For demonstration, we can use a dummy search result structure to test.

# Dummy search results for testing the function
dummy_results = [
    {
        'title': '東京都内で交通事故発生',
        'url': 'http://example.com/news1',
        'content': '昨日、新宿区の交差点付近で交通事故が発生しました。けが人はいませんでした。'
    },
    {
        'title': '渋谷区の事故情報',
        'url': 'http://example.com/news2',
        'content': '渋谷区円山町付近で車両同士の接触事故がありました。'
    },
    {
        'title': '首都高での追突事故',
        'url': 'http://example.com/news3',
        'content': '首都高速3号渋谷線の上り線、用賀出口付近で追突事故により渋滞が発生しています。'
    },
     {
        'title': '練馬区の交通情報',
        'url': 'http://example.com/news4',
        'content': '練馬区内の環七通り沿いで事故による通行止めが発生しました。場所は〇〇です。'
    },
    {
        'title': '特に場所の情報がない記事',
        'url': 'http://example.com/news5',
        'content': '昨日の都内では交通事故が多発しました。安全運転を心がけましょう。'
    }
]

# Test the extraction function with dummy data
# extracted_locations = extract_accident_locations(dummy_results)
# print("\nExtracted Locations (Dummy Data):")
# print(extracted_locations)
```

```python
%%writefile app.py
import streamlit as st
import folium
import os
from tavily import TavilyClient
from datetime import date, timedelta
import re
from geopy.geocoders import Nominatim

# Define functions
def search_traffic_accidents(tavily_client: TavilyClient):
    """
    Uses the Tavily Search API to search for yesterday's traffic accidents in Tokyo.

    Args:
        tavily_client: An initialized TavilyClient object.

    Returns:
        The search results from the Tavily API.
    """
    yesterday = date.today() - timedelta(days=1)
    search_query = f"東京都内の昨日の交通事故 {yesterday.strftime('%Y年%m%d日')} 場所"
    response = tavily_client.search(query=search_query, search_depth="basic")
    return response['results']

def extract_accident_locations(search_results):
    """
    Extracts potential accident locations from Tavily search results.

    Args:
        search_results (list): A list of dictionaries, where each dictionary
                                represents a search result from Tavily.

    Returns:
        list: A list of strings, where each string is a potential accident location.
    """
    locations = []
    location_keywords = ["場所", "付近", "交差点", "区", "市", "町", "村", "丁目", "番地", "地先", "道路"]

    for result in search_results:
        content = result.get('content', '')
        text_to_parse = f"{result.get('title', '')}. {content}"

        pattern = r'(' + '|'.join(location_keywords) + r')\s*([\w\s\d-]+)'
        matches = re.findall(pattern, text_to_parse)

        for keyword, location_text in matches:
            cleaned_location = location_text.strip().replace('。', '').replace('、', '')
            if cleaned_location:
                locations.append(cleaned_location)

    return list(set(locations))

def geocode_locations(locations):
    """
    Geocodes a list of location strings into latitude and longitude coordinates.

    Args:
        locations (list): A list of strings, where each string is a location
                          (address or place name).

    Returns:
        list: A list of dictionaries, where each dictionary contains the original
              location string and its corresponding 'latitude' and 'longitude'.
              Locations that could not be geocoded are excluded.
              (Note: In a real application, you might also want to pass and return date information here if extracted)
    """
    geolocator = Nominatim(user_agent="traffic_accident_mapper", timeout=5)

    geocoded_results = []
    for location_str in locations:
        try:
            search_query = f"{location_str.replace(',', ' ')}, 東京都, 日本"
            location = geolocator.geocode(search_query)
            if location:
                geocoded_results.append({
                    'original_location': location_str,
                    'latitude': location.latitude,
                    'longitude': location.longitude
                })
            else:
                pass
        except Exception as e:
            pass

    return geocoded_results

# Tokyo's approximate coordinates
tokyo_lat = 35.6895
tokyo_lon = 139.6917

# Function to create and display the map with markers
def display_accident_map(geocoded_locations):
    """
    Creates a Folium map of Tokyo and adds markers for geocoded accident locations.

    Args:
        geocoded_locations (list): A list of dictionaries, where each dictionary
                                   contains 'latitude', 'longitude', and
                                   'original_location'.
                                   (Note: Assumes each dict might also contain 'date' if extracted)

    Returns:
        None: Displays the map within the Streamlit application.
    """
    # Create a base map centered on Tokyo
    m = folium.Map(location=[tokyo_lat, tokyo_lon], zoom_start=10)

    # Get yesterday's date to display in the popup (as per the hint)
    yesterday = date.today() - timedelta(days=1)
    accident_date_str = yesterday.strftime('%Y年%m月%d日')


    # Add markers for each geocoded location
    for location_data in geocoded_locations:
        # Construct the popup HTML with location and date
        # If date was extracted and available in location_data, you could use it here:
        # accident_date_display = location_data.get('date', accident_date_str)
        popup_html = f"<b>事故場所:</b> {location_data['original_location']}<br><b>発生日:</b> {accident_date_str}"

        folium.Marker(
            location=[location_data['latitude'], location_data['longitude']],
            popup=popup_html, # Use the HTML string for the popup
            tooltip=location_data['original_location'] # Tooltip shows on hover
        ).add_to(m)

    # Display the map in Streamlit
    import streamlit.components.v1 as components
    map_html = m._repr_html_()
    components.html(map_html, height=500)


# Streamlit App Structure
st.title("東京都 交通事故マップ")

st.write("昨日の東京都内の交通事故情報を検索し、地図上に表示します。")

# Access and initialize Tavily Client
tavily_api_key = os.environ.get("TAVILY_API_KEY")

if not tavily_api_key:
    st.error("TAVILY_API_KEY 環境変数が設定されていません。設定してください。")
else:
    tavily = TavilyClient(api_key=tavily_api_key)

    # Search and Process Accident Data
    st.subheader("事故情報の検索と表示")

    if st.button("昨日の事故を検索"):
        st.info("事故情報を検索しています...")
        search_results = search_traffic_accidents(tavily)

        if search_results:
            st.success("検索が完了しました。")
            accident_locations_str = extract_accident_locations(search_results)

            if accident_locations_str:
                st.info("取得した場所情報をジオコーディングしています...")
                geocoded_locations = geocode_locations(accident_locations_str)

                if geocoded_locations:
                    st.success(f"{len(geocoded_locations)}件の事故場所を特定しました。")

                    # --- Map Display Implementation ---
                    st.subheader("事故発生場所マップ")
                    display_accident_map(geocoded_locations)
                    # --- End Map Display Implementation ---

                else:
                    st.warning("検索結果から事故場所を特定できませんでした。")
            else:
                 st.warning("検索結果から抽出可能な事故場所が見つかりませんでした。")
        else:
            st.warning("交通事故に関する検索結果がありませんでした。")

# Initial message before search
if not st.session_state.get('search_attempted', False):
     st.info("上のボタンをクリックして、昨日の東京都内の交通事故情報を検索してください。")
     st.session_state['search_attempted'] = True
```
