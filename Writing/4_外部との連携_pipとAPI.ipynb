## 4.2. API

外部で提供される機能をプログラミング経由で使うことをＡＰＩと呼びます。ＡＰＩを使うにはＡＰＩ毎にアカウント登録を行って、ＡＰＩキー（もしくはトークン）を取得しなければなりません。ＡＰＩキーを使うことでキーに紐づいたサービスを使うことができます。無料サービスの場合は利用上限などが決まっています。ＡＰＩキーを自分以外の人が知ってしまえば、その人も同じサービスを使うことができ、自分に割り当てられた利用枠が消費されてしまいます。したがってＡＰＩキーは自分以外に対しては秘匿しなければなりません。google colabにはＡＰＩキーを秘匿するための仕組みもあります。

### 4.2.1. APIキーの取得
web検索のためのTavily search apiを使います。  
1. まずTavilyのホームページにアクセスし、アカウント登録し、APIキーを取得（コピー）してください。（無課金です。）
2. 取得したAPIキーをgoogle colabの左側の鍵アイコンをクリックして、  
名前：TAVILY_API_KEY  
値：コピーしたAPIキーをペースト  
ノートブックからのアクセス：チェックを入れる  
としてください。

### 4.2.2. APIキーを環境変数に設定
APIキーを環境変数に設定することで、Geminiからキーにアクセスすることができるようになります。すなわち、Geminiがコーディングの際にWeb検索した情報を使用できるようになります。

```python
# APIキーを環境変数に設定
import os
from google.colab import userdata

os.environ["TAVILY_API_KEY"]=userdata.get('TAVILY_API_KEY')
```
